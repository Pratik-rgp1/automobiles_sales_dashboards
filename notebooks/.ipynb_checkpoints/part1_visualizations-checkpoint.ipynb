```python
#Importing Required Libraries
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
#URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
df = pd.read_csv("../data/historical_automobile_sales.csv")
print('Data read into a dataframe!')
```

# Develop a Line chart using pandas to show how automobile sales fluctuate from year to year

```python
#  Create data for plotting - group by Year and calculate average Automobile Sales
df_line = df.groupby(df['Year'])['Automobile_Sales'].mean()

# Create the line chart
plt.figure(figsize=(10, 6))
df_line.plot(kind='line')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Automobile Sales during Recession')

#  Add x-axis ticks for all years
plt.xticks(ticks=range(1980, 2024), rotation=45)

# Annotate at least two years of recession
plt.text(1981, df_line.loc[1981] + 20, 'Recession 1981', color='red')
plt.text(2009, df_line.loc[2009] + 20, 'Recession 2009', color='red')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Analyze Sales Trends by Vehicle Type During Recession

```python
# Filter dataset to include only recession periods
df_rec = df[df['Recession'] == 1]

# Group by 'Year' and 'Vehicle_Type' to calculate mean automobile sales during recessions
df_Mline = df_rec.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Normalize sales for each vehicle type by dividing sales by the mean sales of that vehicle type
df_Mline['Normalized_Sales'] = df_Mline.groupby('Vehicle_Type')['Automobile_Sales'].transform(lambda x: x / x.mean())

# Set 'Year' as the index for plotting purposes
df_Mline.set_index('Year', inplace=True)

# Initialize plot with size 12x8 inches
plt.figure(figsize=(12, 8))

# Plot normalized sales trends for each vehicle type with markers
for vehicle_type in df_Mline['Vehicle_Type'].unique():
    data = df_Mline[df_Mline['Vehicle_Type'] == vehicle_type]
    plt.plot(data.index, data['Normalized_Sales'], label=vehicle_type, marker='o')

# Highlight recession years with vertical dashed gray lines
recession_years = df_rec['Year'].unique()
for year in recession_years:
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.5)

# Add legend outside plot, axis labels, and title
plt.legend(title="Vehicle Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel("Normalized Sales")
plt.xlabel("Year")
plt.title("Normalized Automobile Sales by Vehicle Type During Recession")

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

<p>Inference:
Sports and supermini cars tend to hold steady or grow during recessions, showing resilience. Medium and small family cars are more affected by economic downturns, with less stable sales. Luxury sports vehicles show strong demand even in tough times.<p/>

# Group the data by 'Recession' and calculate the average Automobile_Sales

```python
new_df = df.groupby('Recession')['Automobile_Sales'].mean().reset_index()

plt.figure(figsize=(10, 6))

# Use the grouped dataframe new_df for plotting
# No need for hue since Recession is on x-axis
sns.barplot(x='Recession', y='Automobile_Sales', data=new_df)

plt.xlabel('Recession Period')
plt.ylabel('Average Automobile Sales')
plt.title('Average Automobile Sales during Recession and Non-Recession')
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])
plt.show()
```

# Group the data by 'Recession' and 'Vehicle_Type' to calculate average Automobile_Sales

```python
grouped_df = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Set figure size for better visualization
plt.figure(figsize=(10, 6))

# Create a grouped bar chart showing average sales by vehicle type during recession and non-recession
sns.barplot(x='Recession', y='Automobile_Sales', hue='Vehicle_Type', data=grouped_df)

# Customize x-axis ticks with meaningful labels
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])

# Label x-axis and y-axis
plt.xlabel('Recession Period')
plt.ylabel('Average Automobile Sales')

# Add title to the plot
plt.title('Vehicle-Wise Sales during Recession and Non-Recession Period')

# Display the plot
plt.show()
```

<p>Inference:
From this plot, we can understand that there is a drastic decline in the overall sales of the automobiles during recession.
However, the most affected type of vehicle is executivecar and sports.<p/>

# Filter data for non-recession years to analyze seasonality impact

```python
non_rec_data = df[df['Recession'] == 0]

# Use 'Seasonality_Weight' column for bubble sizes
size = non_rec_data['Seasonality_Weight']

# Create bubble plot with Month on x-axis and Automobile Sales on y-axis
sns.scatterplot(data=non_rec_data, x='Month', y='Automobile_Sales', size=size, legend=False, sizes=(20, 200))

# Set plot labels and title
plt.xlabel('Month')
plt.ylabel('Automobile_Sales')
plt.title('Seasonality impact on Automobile Sales')

plt.show()
```

```python
# Filter data for recession period
rec_data = df[df['Recession'] == 1]

# Create scatter plot to check correlation between Consumer Confidence and Automobile Sales during recession
plt.scatter(rec_data['Consumer_Confidence'], rec_data['Automobile_Sales'])

# Set axis labels and plot title
plt.xlabel('Consumer Confidence')
plt.ylabel('Automobile Sales')
plt.title('Consumer Confidence and Automobile Sales during Recessions')

plt.show()
```

# Average vehicle price relate to the sales volume during recessions

# Create a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods

```python
# Filter data for recession and non-recession periods
Rdata = df[df['Recession'] == 1]
NRdata = df[df['Recession'] == 0]

# Calculate total advertising expenditure for both periods
RAtotal = Rdata['Advertising_Expenditure'].sum()
NRAtotal = NRdata['Advertising_Expenditure'].sum()

# Create pie chart to show advertising expenditure during recession and non-recession periods
plt.figure(figsize=(8, 6))
labels = ['Recession', 'Non-Recession']
sizes = [RAtotal, NRAtotal]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Advertising Expenditure of XYZAutomotives during Recession and Non-Recession')

plt.show()
```

<p>Inference : 
It seems XYZAutomotives has been spending much more on the advertisements during non-recession periods as compared to during recession times. Fair enough!
<p/>

#  Display the total Advertisement expenditure for each vehicle type during recession period

```python
# Filter data for recession period
Rdata = df[df['Recession'] == 1]

# Calculate total advertising expenditure by vehicle type during recessions
VTexpenditure = Rdata.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

# Create pie chart to show share of each vehicle type in total advertising expenditure during recessions
plt.figure(figsize=(8, 8))

labels = VTexpenditure.index
sizes = VTexpenditure.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Advertisement Expenditure Share by Vehicle Type during Recession')

plt.show()
```

<p>Inference : 
During recession the advertisements were mostly focued on low price range vehicle. A wise decision!
<p/>

# Effect of Unemployment Rate on Vehicle Type and Sales during Recession

```python
# Filter data for recession period
df_rec = df[df['Recession'] == 1]

# Create lineplot to analyze unemployment rate effect on vehicle sales by vehicle type
sns.lineplot(
    data=df_rec,
    x='unemployment_rate',
    y='Automobile_Sales',
    hue='Vehicle_Type',
    style='Vehicle_Type',
    markers='o',
    err_style=None
)

plt.ylim(0, 850)
plt.legend(loc=(0.05, 0.3))
plt.title('Effect of Unemployment Rate on Vehicle Type and Sales')

plt.show()
# Inference:
# Sales of supermini cars, small family cars, and medium family cars tend to decrease as the unemployment rate increases,
# indicating their sensitivity to economic conditions during recessions. Supermini cars show more resilience compared to others.
```
